### TASK 1-->Create a pandas dataframe (DataFrame name as 'df') with numpy random values (4 features and 4 observation)


```python
import pandas as pd
import numpy as np


np.random.seed(42)


data = np.random.randn(4, 4)

df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])


print(df)
```
